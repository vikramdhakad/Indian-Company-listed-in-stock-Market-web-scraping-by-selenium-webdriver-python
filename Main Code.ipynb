```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from time import sleep
from webdriver_manager.chrome import ChromeDriverManager
import requests
```

```python
chrome_options = Options()
chrome_options.add_experimental_option("detach",True)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=chrome_options)
driver.get('https://www.csr.gov.in/content/csr/global/master/home/home/csr-expenditure--geographical-distribution/companies.html?=FY%202020-21')
sleep(10)
driver.maximize_window()
#driver.find_element(By.XPATH,'//select[@class="total-pages"]').click()
#driver.find_element(By.XPATH,'//option[@value="20"]').click()
#sleep(5)
```

```python
for j in range(2,4):
    names = []
    amount = []
    classes = []
    stat = []
    com_type = []
    roc_t = []
    sub_ct = []
    statu = []
    avg_pro = []
    expend = []
    spt = []
    local_spt = []
    
    for com in range(1,11):
        co = driver.find_element(By.XPATH,'//*[@id="compTableBody"]/tr[{}]/td[2]/span'.format(com))
        co.click()
        sleep(10)
        driver.switch_to.window(driver.window_handles[1])
        sleep(2)
        
        try:
            name = driver.find_element(By.XPATH,'//*[@id="headingText"]')
            names.append(name.text) 
        except:
            names = " "
        
        try:
            amount_in_cr = driver.find_element(By.XPATH,'//*[@id="csrDetailsTableBody"]/tr[21]/td[6]/div')
            amount.append(amount_in_cr.text)   
        except:
            amount = " "
            
        try:
            clas = driver.find_element(By.XPATH,'//span[@id="classOfCompany"]')
            classes.append(clas.text)
        except:
            classes = []
        
        try:
            state = driver.find_element(By.XPATH,'//span[@id="stateOfCompany"]')
            stat.append(state.text)
        except:
            stat = " "
            
        try:
            company_type = driver.find_element(By.XPATH,'//span[@id="typeOfCompany"]')
            com_type.append(company_type.text)
        except:
            com_type = " "
            
        try:
            roc = driver.find_element(By.XPATH,'//span[@id="rocOfCompany"]')
            roc_t.append(roc.text)
        except:
            roc_t = " "
            
        try:
            sub_cat = driver.find_element(By.XPATH,'//span[@id="categoryOfCompany"]')
            sub_ct.append(sub_cat.text)
        except:
            sub_ct = " "
            
        try:
            status = driver.find_element(By.XPATH,'//span[@id="listingStatusOfCompany"]')
            statu.append(status.text)
        except:
            statu = " "
            
        try:
            avg_profit = driver.find_element(By.XPATH,'//span[@id="averageProfit"]')
            avg_pro(avg_profit.text)
        except:
            avg_pro = " "
        
        try:
            expenditure = driver.find_element(By.XPATH,'//span[@id="prescribedExp"]')
            expend.append(expenditure.text)
        except:
            expend = " "
        
        try:
            spent = driver.find_element(By.XPATH,'//span[@id="csrSpent"]')
            spt.append(spent.text)
        except:
            spt = " "
            
        try: 
            local_spent = driver.find_element(By.XPATH,'//span[@id="localSpent"]')
            local_spt.append(local_spent.text)
        except:
            local_spt = " "
            
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        
    pages = driver.find_element(By.LINK_TEXT,'{}'.format(j))
    pages.click()
    sleep(15)
```

```python
data = pd.DataFrame()
data['Company Name'] = names
data['Amount in Cr'] = amount
data['Class'] = classes
data['State'] = stat
data['Company Type'] = com_type
data['ROC'] = roc_t
data['Sub Category'] = sub_ct
data['Status'] = statu
data['Average Net Profit'] = avg_pro
data['CSR Prescribed Expenditure'] = expend
data['CSR Spent'] = spt
data['Local Area Spent'] = local_spt
```

```python
data.to_excel('Sample.xlsx')
```
